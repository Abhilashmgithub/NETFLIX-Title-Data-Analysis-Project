**Data Visualization**

```python
#Taking count of rating available
data.groupby(['rating']).size().reset_index(name='counts')
```

```python
#Creating the Pie chart based on content rating
df = pd.DataFrame(data.head(50))
df['rating'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title("Distribution of content ratings on Netflix")
plt.show()
```

```python
# It Shows Which country release most movies in each year?
data.country.value_counts()
```

```python
#Creating a bar chart to show top 10 countries release show on Netflix
data.country.value_counts().head(10).plot(kind='bar',color='purple')
plt.title('Top 10 countries release show on Netflix')
plt.xlabel("country")
plt.ylabel('counts')
plt.show()
```

```python
#Top Geners listed in netflix
data.listed_in.value_counts()
```

```python
#Showing the Top 10 genres watched on Netflix
data.listed_in.value_counts().head(10).plot(kind='barh',color="g")
plt.title("Top 10 Geners watched on Netflix")
plt.ylabel('Geners')
plt.legend()
```

```python
#Types of shows in Netflix
data.type.value_counts()
```

```python
#Creating a pie chart based on content of Types in Netflix
data.type.value_counts().plot(kind='pie')
plt.title("Most watched show type on Netflix")
plt.legend()
plt.show()
```

```python
# Most movies released in year
data.release_year.value_counts().head(20)
```

```python
#How many movies are released in each year?
data.release_year.value_counts().head(20).plot(kind='barh',color='yellow')
plt.title('Top 10 years in which most movies are released in Netflix')
plt.xlabel("Release Year")
plt.ylabel('counts')
plt.legend()
plt.show()
```

```python
#Most number of movies directed by The Directors.
data.director.value_counts().head(10)
```

```python
data.director.value_counts().head(10).plot(kind='bar', color='orange')
plt.title('Most movies directed by The Directors.')
plt.legend()
```

```python
#maximum number of dates added.
data.date_added.value_counts().head(5)
```

```python
data.date_added.value_counts().head(5).plot(linestyle='dashed')
plt.title("Date Added")
plt.legend()
plt.xticks(rotation=60)
plt.show()
```

```python
#Maximum Duration of movies/Tv shows
data.duration.value_counts().head(10)
```

```python

data.duration.value_counts().head(10).plot(kind='bar', color='Blue')
plt.xlabel('Duration of Movies/ Tv shows')
plt.ylabel('count')
plt.title("Maximum Duration of Movies / Tv shows")
plt.show()
```
